```python
coluna = []

for i in range(0, len(X[i])):
    
    for j in range(0, len(X)):
        coluna.append(X[j][i])
    plt.figure(i)
    
    plt.grid()
    plt.scatter(coluna, y)
    del coluna[:]
```

### Questão 3

```python
coluna = []
coeficientes = []
for i in range(0, len(X[i])):
    
    for j in range(0, len(X)):
        coluna.append(X[j][i])
    coeficientes.append(np.corrcoef(coluna, y))
    del coluna[:]
```

```python
print(coeficientes)
```
