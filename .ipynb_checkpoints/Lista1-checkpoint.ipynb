### Questão 2

```python
coluna = []

for i in range(0, len(X[i])):
    
    for j in range(0, len(X)):
        coluna.append(X[j][i])
    plt.figure(i)
    
    plt.grid()
    plt.scatter(coluna, y)
    del coluna[:]
```

### Questão 3

```python
coluna = []
coeficientes = []
for i in range(0, len(X[i])):
    
    for j in range(0, len(X)):
        coluna.append(X[j][i])
    coeficientes.append(np.corrcoef(coluna, y))
    del coluna[:]
```

```python
model_8.fit(XMelhor, y)
```

```python
ymelhor_pred = model_8.predict(XMelhor)
```

```python
ymelhor_pred
```

### Automático

```python
model_8.coef_
```

```python
model_8.intercept_
```

### Manual

```python
coef = [ [i] for i in y ]
coef
```

### B1


```python
b1 = np.sum((XMelhor - np.mean(XMelhor)) * (coef - np.mean(coef))) / np.sum((XMelhor - np.mean(XMelhor))**2)
```

```python
b1
```

### B0


```python
b0 = np.mean(coef) - b1 * np.mean(X)
```

```python
b0
```

### Gráfico

```python
plt.grid()
plt.scatter(XMelhor, y, color='red')
plt.plot(XMelhor, ymelhor_pred)
plt.show()
```

### MSE
